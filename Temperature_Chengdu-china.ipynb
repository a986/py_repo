# Observing Temperatures 

#### Observing the temperatures of China's Chengdu area precisely PM-2.5 affecting humidity and temperature.

## Humidity Vs Temperature 

```python
# Importing the major libraries required for exploration
import pandas as pd
from matplotlib import  pyplot as plt

#setting the file path
file_path = "./PM2.5/ChengduPM20100101_20151231.csv"
```

```python
# inserting the dataset into df(dataframe)
df = pd.read_csv(file_path)


period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
df["datetime"] = period
print(df.head(10))

```

```python
#Inserting the Date and time
df.set_index("datetime",inplace=True)


df = df.resample("7D").mean()
print(df.head())
```

```python
#observing the date and time
data  =df["HUMI"] #DEWP,HUMI
data_china = df["TEMP"]

print(data_china.head(100))
```

```python
_x = data.index
_x = [i.strftime("%Y-%m-%d") for i in _x]
_x_china = [i.strftime("%Y-%m-%d") for i in data_china.index]
print(len(_x_china),len(_x_china))
_y = data.values
_y_china = data_china.values
```

```python
#plotting the graph to see the trends
plt.figure(figsize=(20,8),dpi=80)

#Labelling the axis
plt.plot(range(len(_x)),_y,label="HUMI",alpha=0.7)
plt.plot(range(len(_x_china)),_y_china,label="TEMP",alpha=0.7)


#inserting lengths and marks for X - axis
plt.xticks(range(0,len(_x_china),10),list(_x_china)[::10],rotation=45)

#defining the legend
plt.legend(loc="best")

#displaying the plot
plt.show()
```

### Conclusion - 
#### Observing the above graph we found that the temperature though at a steady circular graph, Humidity can be seen at a vast level. Here we can see humidity bouncing randomnly with no pattern and getting higher.

## Dew Precipitation vs Temperature

```python
df = pd.read_csv(file_path)


period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
df["datetime"] = period
print(df.head(10))
```

```python
#Setting the date and time and setting it on index
df.set_index("datetime",inplace=True)

df = df.resample("7D").mean()
print(df.head())
```

```python
#PRES,TEMP
data  =df["DEWP"]  
data_china = df["TEMP"]

print(data_china.head(100))
```

```python
_x = data.index
_x = [i.strftime("%Y-%m-%d") for i in _x]
_x_china = [i.strftime("%Y-%m-%d") for i in data_china.index]
print(len(_x_china),len(_x_china))
_y = data.values
_y_china = data_china.values
```

```python
plt.figure(figsize=(20,8),dpi=80)

plt.plot(range(len(_x)),_y,label="DEWP",alpha=0.7)
plt.plot(range(len(_x_china)),_y_china,label="Temperature",alpha=0.7)

plt.xticks(range(0,len(_x_china),10),list(_x_china)[::10],rotation=45)

plt.legend(loc="best")

plt.show()
```

### Conclusion -
#### Here we can see that the dew in precipitation with respect to temperature remains close but the humidity has spike which is affecting the climate in this area.
