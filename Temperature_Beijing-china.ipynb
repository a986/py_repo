# Observing Temperatures

#### Observing the temperatures of China's Beijing area precisely PM-2.5 affecting humidity and temperature.

## Humidity Vs Temperature

```python
# Importing the nexessary Libraries.
import pandas as pd
from matplotlib import  pyplot as plt

#setting the file path
file_path = "./PM2.5/BeijingPM20100101_20151231.csv"
```

```python
#defining the dataset as df and inserting the values in it.
df = pd.read_csv(file_path)

#Giving the index to the year days and months
period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
df["datetime"] = period
print(df.head(10))

```

```python
#setting the date and time
df.set_index("datetime",inplace=True)


df = df.resample("7D").mean()
print(df.head())
```

```python
data  =df["HUMI"]
data_china = df["TEMP"]

print(data_china.head(100))
```

```python
_x = data.index
_x = [i.strftime("%Y-%m-%d") for i in _x]
_x_china = [i.strftime("%Y-%m-%d") for i in data_china.index]
print(len(_x_china),len(_x_china))
_y = data.values
_y_china = data_china.values
```

```python
plt.figure(figsize=(20,8),dpi=80)

plt.plot(range(len(_x)),_y,label="Humidity",alpha=0.7)
plt.plot(range(len(_x_china)),_y_china,label="Temperature",alpha=0.7)

plt.xticks(range(0,len(_x_china),10),list(_x_china)[::10],rotation=45)

plt.legend(loc="best")

plt.show()
```

## Conclusion -
#### On observing the graphs, it can be concluded that temperatures are in sync with humidity but slightly higher. It keeps on dropping and rising according to weather.

## Dew Precipitation vs Temperature

```python
df = pd.read_csv(file_path)


period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
df["datetime"] = period
print(df.head(10))
```

```python
#Setting the date and time and setting it on index
df.set_index("datetime",inplace=True)

df = df.resample("7D").mean()
print(df.head())
```

```python
data  =df["DEWP"]  
data_china = df["TEMP"]

print(data_china.head(100))
```

```python
_x = data.index
_x = [i.strftime("%Y-%m-%d") for i in _x]
_x_china = [i.strftime("%Y-%m-%d") for i in data_china.index]
print(len(_x_china),len(_x_china))
_y = data.values
_y_china = data_china.values
```

```python
plt.figure(figsize=(20,8),dpi=80)

plt.plot(range(len(_x)),_y,label="DEW-Precipitation",alpha=0.7)
plt.plot(range(len(_x_china)),_y_china,label="Temperature",alpha=0.7)

plt.xticks(range(0,len(_x_china),10),list(_x_china)[::10],rotation=45)

plt.legend(loc="best")

plt.show()
```

## Conclusion
#### Here again the graph of tempertature and dew precipitation are in sync. So can be concluded for having decent weather.
